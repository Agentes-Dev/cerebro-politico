Teste

```python
# Passo 1: Obter dados de deputados
deputados = []
for id_deputado in df['id']:
    url_deputado = f'https://dadosabertos.camara.leg.br/api/v2/deputados/{id_deputado}'
    r_deputado = requests.get(url_deputado).json()
    deputado_info = pd.json_normalize(r_deputado['dados'])[['nomeCivil', 'cpf','sexo', 'dataNascimento', 'dataFalecimento', 'ufNascimento', 'escolaridade', 'ultimoStatus.id', 'ultimoStatus.siglaPartido', 'ultimoStatus.idLegislatura', 'ultimoStatus.gabinete.telefone', 'ultimoStatus.situacao', 'ultimoStatus.condicaoEleitoral',
       'ultimoStatus.descricaoStatus']]
    deputados.append(deputado_info)

# Passo 2: Obter dados de despesas de cada deputado
despesas = []
for id_deputado in df['id']:
    url_despesas = f'https://dadosabertos.camara.leg.br/api/v2/deputados/{id_deputado}/despesas'
    r_despesas = requests.get(url_despesas).json()
    df_despesas = pd.DataFrame(r_despesas['dados'])
    df_despesas['id_deputado'] = id_deputado
    despesas.append(df_despesas)

# Passo 3: Criar o DataFrame final com informações de deputados
deputados_df = pd.concat(deputados, ignore_index=True)

df = pd.merge(df, deputados_df, left_on='id', right_on='ultimoStatus.id')

# Lista de IDs dos deputados
ids_deputados = df['id'].tolist()# Dicionários para armazenar informações

eventos_dict = {}
frentes_dict = {}
mandatos_dict = {}
ocupacoes_dict = {}
orgaos_dict = {}
profissoes_dict = {}

# Realize as requisições para cada deputado e armazene os dados nos dicionários
for id_deputado in ids_deputados:
    # Informações de Eventos Parlamentares
    url_eventos = f'https://dadosabertos.camara.leg.br/api/v2/deputados/{id_deputado}/eventos'
    r_eventos = requests.get(url_eventos).json()
    eventos_dict[id_deputado] = r_eventos['dados']

    # Informações de Frentes Parlamentares
    url_frentes = f'https://dadosabertos.camara.leg.br/api/v2/deputados/{id_deputado}/frentes'
    r_frentes = requests.get(url_frentes).json()
    frentes_dict[id_deputado] = r_frentes['dados']

    # Informações de Mandatos Externos
    url_mandatosExternos = f'https://dadosabertos.camara.leg.br/api/v2/deputados/{id_deputado}/mandatosExternos'
    r_mandatosExternos = requests.get(url_mandatosExternos).json()
    mandatos_dict[id_deputado] = r_mandatosExternos['dados']

    # Informações de Ocupações
    url_ocupacoes = f'https://dadosabertos.camara.leg.br/api/v2/deputados/{id_deputado}/ocupacoes'
    r_ocupacoes = requests.get(url_ocupacoes).json()
    ocupacoes_dict[id_deputado] = r_ocupacoes['dados']

    # Informações de Órgãos
    url_orgaos = f'https://dadosabertos.camara.leg.br/api/v2/deputados/{id_deputado}/orgaos'
    r_orgaos = requests.get(url_orgaos).json()
    orgaos_dict[id_deputado] = r_orgaos['dados']

    # Informações de Profissões
    url_profissoes = f'https://dadosabertos.camara.leg.br/api/v2/deputados/{id_deputado}/profissoes'
    r_profissoes = requests.get(url_profissoes).json()
    profissoes_dict[id_deputado] = r_profissoes['dados']

# Adicione as colunas ao DataFrame final
df['eventos'] = df['id'].map(eventos_dict)
df['frentes'] = df['id'].map(frentes_dict)
df['mandatos_externos'] = df['id'].map(mandatos_dict)
df['ocupacoes'] = df['id'].map(ocupacoes_dict)
df['orgaos'] = df['id'].map(orgaos_dict)
df['profissoes'] = df['id'].map(profissoes_dict)


# Passo 4: Criar o DataFrame final
despesas_df = pd.concat(despesas, ignore_index=True)

# Agora você pode combinar os DataFrames de despesas e informações do deputado com base no ID do deputado
df_final = pd.merge(df, despesas_df, left_on='id', right_on='id_deputado')

# df_final agora contém as informações dos deputados e as despesas em um único DataFrame
df_final
```
