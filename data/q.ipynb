Líderes

```python
url_lideres = 'https://dadosabertos.camara.leg.br/api/v2/legislaturas/57/lideres'
r_lideres = requests.get(url_lideres).json()

#Criando colunas normalizadas
df_lideres = pd.DataFrame(r_lideres['dados'])
df_lideres['idParlamentar'] = df_lideres['parlamentar'].apply(lambda x: x['id'])
df_lideres['nome'] = df_lideres['parlamentar'].apply(lambda x: x['nome'])
df_lideres['siglaPartido'] = df_lideres['parlamentar'].apply(lambda x: x['siglaPartido'])
df_lideres['siglaUf'] = df_lideres['parlamentar'].apply(lambda x: x['siglaUf'])
df_lideres['idLegislatura'] = df_lideres['parlamentar'].apply(lambda x: x['idLegislatura'])
df_lideres['tipoLider'] = df_lideres['bancada'].apply(lambda x: x['tipo'])
df_lideres['tipoNome'] = df_lideres['bancada'].apply(lambda x: x['nome'])

df_lideres.drop(columns=['parlamentar', 'bancada'], inplace=True)

df_lideres

```
